```python
#!/usr/bin/env python3
import math
import sys
import os
import itertools
import random
import copy

##############################
# Utility Functions
##############################

def euclidean_distance(item1, item2):
    """
    Compute Euclidean distance between two items.
    (Each item is a tuple: (value, weight).)
    Used only for ordering in the greedy construction.
    """
    return math.sqrt((item1[0] - item2[0])**2 + (item1[1] - item2[1])**2)

def total_value(solution, items):
    """Return the total value (X) of the items in the solution.
       The solution is represented as a set of item indices."""
    return sum(items[i][0] for i in solution)

def total_weight(solution, items):
    """Return the total weight (Y) of the items in the solution."""
    return sum(items[i][1] for i in solution)

def is_feasible(solution, items, capacity):
    """Return True if the solution (a set of item indices) is feasible."""
    return total_weight(solution, items) <= capacity

def better_solution(sol_a, sol_b, items):
    """
    Decide if solution A is better than solution B.
    A solution is considered better if its total value is higher,
    or if total values are equal and its total weight is lower.
    Both sol_a and sol_b are assumed feasible.
    """
    va, wa = total_value(sol_a, items), total_weight(sol_a, items)
    vb, wb = total_value(sol_b, items), total_weight(sol_b, items)
    if va > vb:
        return True
    elif va == vb and wa < wb:
        return True
    return False

##############################
# Instance Loader
##############################

def load_knapsack_instance(file_path):
    """
    Load a knapsack instance from file.

    Expected file format:
      - First line: number_of_items capacity
      - Each subsequent line: value weight
    Returns:
      capacity: float
      items: list of tuples (value, weight)
    """
    items = []
    with open(file_path, "r") as f:
        lines = [line.strip() for line in f if line.strip()]
    first_line = lines[0].split()
    if len(first_line) < 2:
        raise ValueError("First line must contain number of items and capacity")
    n = int(first_line[0])
    capacity = float(first_line[1])
    for line in lines[1:]:
        parts = line.split()
        if len(parts) < 2:
            continue
        value = float(parts[0])
        weight = float(parts[1])
        items.append((value, weight))
    if len(items) != n:
        print(f"Warning: Expected {n} items but found {len(items)} items.")
    return capacity, items

##############################
# Greedy Construction (for seeding GA)
##############################

def longest_neighbor_route(items, capacity, start=None):
    """
    Build an initial feasible solution using a greedy (longest-neighbor) heuristic.

    If start is None, choose the item with maximum value.
    Then repeatedly add the item (from those not yet selected) that is the farthest
    (by Euclidean distance) from the last added item and that does not exceed capacity.

    Returns:
      A list of selected item indices.
    """
    n = len(items)
    if start is None:
        start = max(range(n), key=lambda i: items[i][0])
    solution = [start]
    used = set(solution)
    current_weight = items[start][1]

    while True:
        current_item = items[solution[-1]]
        candidate = None
        best_dist = -1
        for i in range(n):
            if i in used:
                continue
            if current_weight + items[i][1] > capacity:
                continue
            d = euclidean_distance(current_item, items[i])
            if d > best_dist:
                best_dist = d
                candidate = i
        if candidate is None:
            break
        solution.append(candidate)
        used.add(candidate)
        current_weight += items[candidate][1]
    return solution

##############################
# Local Improvement: Inverted opt4
##############################

def inverted_opt4(initial_solution, items, capacity, max_multi_swap=3):
    """
    Aggressively improve a given knapsack solution using a variety of moves.

    Moves include:
      1. Insertion Moves: Add an item not in the solution.
      2. Single Swap Moves: Replace one item with another.
      3. Multi-Swap Moves: Remove k items and add k different items (for 2 <= k <= max_multi_swap).
      4. Flip Moves: Toggle the inclusion state of each item.
      5. Optional Removal Moves: Remove items if that reduces weight while keeping value.

    A move is accepted if the new solution is feasible and is "better" (higher total value;
    if equal, lower total weight).

    The process is repeated until no further improvement is found.

    Parameters:
      initial_solution : list (or set) of item indices (initial solution)
      items            : list of items (value, weight)
      capacity         : knapsack capacity
      max_multi_swap   : maximum number of items to exchange simultaneously (default 3)

    Returns:
      A sorted list of item indices representing the improved solution.
    """
    current_solution = set(initial_solution)
    improved = True

    while improved:
        improved = False
        best_solution = current_solution  # best in this iteration

        outside = set(range(len(items))) - current_solution

        # 1. Insertion Moves
        for i in list(outside):
            new_sol = current_solution | {i}
            if is_feasible(new_sol, items, capacity) and better_solution(new_sol, best_solution, items):
                best_solution = new_sol
                improved = True
                break
        if improved:
            current_solution = best_solution
            continue

        # 2. Single Swap Moves
        for a in list(current_solution):
            for b in list(outside):
                new_sol = (current_solution - {a}) | {b}
                if is_feasible(new_sol, items, capacity) and better_solution(new_sol, best_solution, items):
                    best_solution = new_sol
                    improved = True
                    break
            if improved:
                break
        if improved:
            current_solution = best_solution
            continue

        # 3. Multi-Swap Moves: Try swapping sets of k items.
        for k in range(2, max_multi_swap+1):
            if len(current_solution) < k or len(outside) < k:
                continue
            for remove_set in itertools.combinations(current_solution, k):
                for add_set in itertools.combinations(outside, k):
                    new_sol = (current_solution - set(remove_set)) | set(add_set)
                    if not is_feasible(new_sol, items, capacity):
                        continue
                    if better_solution(new_sol, best_solution, items):
                        best_solution = new_sol
                        improved = True
                        break
                if improved:
                    break
            if improved:
                break
        if improved:
            current_solution = best_solution
            continue

        # 4. Flip Moves: Toggle inclusion for each item.
        for i in range(len(items)):
            if i in current_solution:
                new_sol = current_solution - {i}
            else:
                new_sol = current_solution | {i}
            if is_feasible(new_sol, items, capacity) and better_solution(new_sol, best_solution, items):
                best_solution = new_sol
                improved = True
                break
        if improved:
            current_solution = best_solution
            continue

        # 5. Optional Removal Moves: Remove an item if it reduces weight without lowering total value.
        for a in list(current_solution):
            new_sol = current_solution - {a}
            if is_feasible(new_sol, items, capacity) and total_value(new_sol, items) == total_value(current_solution, items):
                if total_weight(new_sol, items) < total_weight(current_solution, items):
                    best_solution = new_sol
                    improved = True
                    break
        if improved:
            current_solution = best_solution
            continue

    return sorted(list(current_solution))

##############################
# Genetic Algorithm Operators
##############################

def individual_to_bitstring(individual, n):
    """Convert a set (or list) of item indices to a bitstring (list of 0/1) of length n."""
    bitstring = [0] * n
    for i in individual:
        bitstring[i] = 1
    return bitstring

def bitstring_to_individual(bitstring):
    """Convert a bitstring (list of 0/1) to a set of item indices."""
    return {i for i, bit in enumerate(bitstring) if bit}

def fitness(individual, items, capacity):
    """
    Fitness function: if feasible, return total value; otherwise, return a very low fitness.
    (We assume that infeasible solutions are heavily penalized.)
    """
    if is_feasible(individual, items, capacity):
        return total_value(individual, items)
    else:
        return -1e6  # very low fitness for infeasible individuals

def tournament_selection(population, items, capacity, tournament_size=3):
    """Perform tournament selection on the population."""
    selected = []
    pop_size = len(population)
    for _ in range(pop_size):
        tournament = random.sample(population, tournament_size)
        tournament.sort(key=lambda ind: fitness(ind, items, capacity), reverse=True)
        selected.append(copy.deepcopy(tournament[0]))
    return selected

def uniform_crossover(parent1, parent2, n):
    """
    Perform uniform crossover between two parents (bitstring representation).
    Returns two children (as bitstrings).
    """
    child1 = [0]*n
    child2 = [0]*n
    for i in range(n):
        if random.random() < 0.5:
            child1[i] = parent1[i]
            child2[i] = parent2[i]
        else:
            child1[i] = parent2[i]
            child2[i] = parent1[i]
    return child1, child2

def mutation(bitstring, mutation_rate):
    """Bit-flip mutation on a bitstring."""
    n = len(bitstring)
    new_bitstring = bitstring[:]
    for i in range(n):
        if random.random() < mutation_rate:
            new_bitstring[i] = 1 - new_bitstring[i]
    return new_bitstring

def repair_and_improve(bitstring, items, capacity, max_multi_swap=3):
    """
    Convert the bitstring into a solution, repair it using local improvement (inverted_opt4)
    to ensure feasibility and push toward high value, and return the improved bitstring.
    """
    individual = bitstring_to_individual(bitstring)
    # If infeasible, try to remove items (greedy removal)
    if not is_feasible(individual, items, capacity):
        sorted_items = sorted(list(individual), key=lambda i: items[i][0]/items[i][1])  # low bang-for-buck first
        for i in sorted_items:
            individual.remove(i)
            if is_feasible(individual, items, capacity):
                break
    # Improve locally using inverted opt4
    improved = inverted_opt4(list(individual), items, capacity, max_multi_swap=max_multi_swap)
    return individual_to_bitstring(improved, len(items))

##############################
# Genetic Algorithm Main Function
##############################

def genetic_algorithm(items, capacity, pop_size=50, generations=100, crossover_rate=0.8, mutation_rate=0.05, max_multi_swap=3):
    n = len(items)
    population = []

    # --- Initialization ---
    # We initialize the population with a mix of greedy solutions and random bitstrings.
    # First, seed with greedy solutions (using various starting points).
    for start in range(min(n, pop_size//2)):
        init_solution = set(longest_neighbor_route(items, capacity, start))
        # Repair if needed and improve locally.
        bit = repair_and_improve(individual_to_bitstring(init_solution, n), items, capacity, max_multi_swap)
        population.append(bit)
    # Fill the remainder of the population with random feasible individuals.
    while len(population) < pop_size:
        bit = [0] * n
        indices = list(range(n))
        random.shuffle(indices)
        current = set()
        for i in indices:
            if total_weight(current | {i}, items) <= capacity:
                current.add(i)
        bit = individual_to_bitstring(current, n)
        bit = repair_and_improve(bit, items, capacity, max_multi_swap)
        population.append(bit)

    best_individual = None
    best_fitness = -1e9

    # --- Main GA Loop ---
    for gen in range(generations):
        # Evaluate fitness and record best individual.
        pop_fitness = [fitness(bitstring_to_individual(ind), items, capacity) for ind in population]
        for i, fit_val in enumerate(pop_fitness):
            if fit_val > best_fitness:
                best_fitness = fit_val
                best_individual = population[i]
        print(f"Generation {gen}: Best fitness = {best_fitness:.4f}")

        # Selection
        selected = tournament_selection(population, items, capacity, tournament_size=3)

        # Crossover
        next_population = []
        for i in range(0, pop_size-1, 2):
            parent1 = selected[i]
            parent2 = selected[i+1]
            if random.random() < crossover_rate:
                child1, child2 = uniform_crossover(parent1, parent2, n)
            else:
                child1, child2 = parent1[:], parent2[:]
            next_population.extend([child1, child2])
        if len(next_population) < pop_size:
            next_population.append(selected[-1])

        # Mutation and local improvement (repair)
        new_population = []
        for ind in next_population:
            mutated = mutation(ind, mutation_rate)
            repaired = repair_and_improve(mutated, items, capacity, max_multi_swap)
            new_population.append(repaired)
        population = new_population

    # Convert the best individual (bitstring) to a solution (set of indices)
    best_solution = bitstring_to_individual(best_individual)
    return best_solution

##############################
# Main Execution
##############################

def main():
    # Default instance file (change as needed)
    default_instance = "/content/knapsack-01-instances-main/pisinger_instances_01_KP/low-dimensional/f10_l-d_kp_20_879"

    # Read instance file from command-line if provided.
    if len(sys.argv) < 2 or sys.argv[1] in ["-f", "--file"]:
        instance_file = default_instance
        print(f"No valid instance file provided. Using default: {instance_file}")
    else:
        instance_file = sys.argv[1]

    if not os.path.exists(instance_file):
        print(f"Instance file '{instance_file}' not found!")
        sys.exit(1)

    capacity, items = load_knapsack_instance(instance_file)
    n = len(items)
    print(f"Loaded instance with capacity = {capacity} and {n} items.")

    # Run the genetic algorithm.
    best_solution = genetic_algorithm(
        items,
        capacity,
        pop_size=50,
        generations=100,
        crossover_rate=0.8,
        mutation_rate=0.05,
        max_multi_swap=3
    )

    best_value = total_value(best_solution, items)
    best_weight = total_weight(best_solution, items)

    print("\nFinal Best Solution Found by GA:")
    print(f"  Selected {len(best_solution)} out of {n} items")
    print(f"  Total Value = {best_value:.4f}")
    print(f"  Total Weight = {best_weight:.4f} (Capacity = {capacity:.4f})")
    print(f"  Items (indices): {sorted(list(best_solution))}")

if __name__ == '__main__':
    main()
```

```python
import heapq
import numpy as np

def load_knapsack_instance(file_path):
    """
    Loads a knapsack instance in the Pisinger format.

    Expected file format:
      - The first line contains two integers: n (number of items) and capacity.
      - The next n lines each contain two integers: value and weight for each item.

    Returns:
      values (list of int): The list of item values.
      weights (list of int): The list of item weights.
      capacity (int): The knapsack capacity.
    """
    with open(file_path, "r") as f:
        lines = f.readlines()

    if not lines:
        raise ValueError("The file is empty.")

    try:
        n, capacity = map(int, lines[0].split())
    except Exception as e:
        raise ValueError("The first line must contain two integers: number of items and capacity.") from e

    if len(lines) < n + 1:
        raise ValueError(f"Expected {n} item lines, but only found {len(lines)-1}.")

    values, weights = [], []
    for i in range(1, n + 1):
        try:
            v, w = map(int, lines[i].split())
        except Exception as e:
            raise ValueError(f"Line {i+1} is not in the expected 'value weight' format.") from e
        values.append(v)
        weights.append(w)
    return values, weights, capacity

def turbo_boosted_knapsack_solver(values, weights, capacity):
    """
    Turbo-boosted Branch and Bound solver for the 0-1 Knapsack Problem.

    Enhancements:
      - Items are sorted by value-to-weight ratio (descending).
      - Precomputed cumulative sums allow a fast (binary search based)
        calculation of a fractional (greedy) upper bound.
      - A best-first search using a priority queue (max-heap) is employed.

    Parameters:
      values: list of item values.
      weights: list of item weights.
      capacity: maximum weight allowed.

    Returns:
      best_value: optimal total value.
      best_solution: binary list (in original order) indicating selected items.
    """
    n = len(values)

    # Sort items by descending ratio (value/weight)
    sorted_indices = sorted(range(n), key=lambda i: values[i] / weights[i], reverse=True)
    sorted_values = [values[i] for i in sorted_indices]
    sorted_weights = [weights[i] for i in sorted_indices]

    # Precompute cumulative sums for the sorted order.
    # cumulative_weight[j] = total weight of items 0 to j-1 in the sorted order.
    # cumulative_value[j]  = total value of items 0 to j-1 in the sorted order.
    cumulative_weight = np.zeros(n + 1)
    cumulative_value = np.zeros(n + 1)
    for j in range(n):
        cumulative_weight[j + 1] = cumulative_weight[j] + sorted_weights[j]
        cumulative_value[j + 1] = cumulative_value[j] + sorted_values[j]

    def compute_bound(index, curr_weight, curr_value):
        """
        Compute an upper bound on the total value that can be achieved,
        starting from the current branch (with items [index, n-1] in the sorted order).

        This function uses binary search on the precomputed cumulative arrays.
        """
        remaining = capacity - curr_weight
        if remaining <= 0:
            return curr_value
        # Find the largest k (>= index) such that:
        # cumulative_weight[k] - cumulative_weight[index] <= remaining
        low = index
        high = n
        while low < high:
            mid = (low + high + 1) // 2
            if cumulative_weight[mid] - cumulative_weight[index] <= remaining:
                low = mid
            else:
                high = mid - 1
        bound_value = curr_value + (cumulative_value[low] - cumulative_value[index])
        # Take a fraction of the next item, if any.
        if low < n:
            remaining_fraction = remaining - (cumulative_weight[low] - cumulative_weight[index])
            if sorted_weights[low] > 0:
                fraction = remaining_fraction / sorted_weights[low]
                bound_value += fraction * sorted_values[low]
        return bound_value

    # Priority queue node: (negative_bound, current_value, current_weight, index, taken_items)
    # We use negative_bound because heapq in Python is a min-heap.
    initial_bound = compute_bound(0, 0, 0)
    pq = [(-initial_bound, 0, 0, 0, [])]

    best_value = 0
    best_taken_sorted = []  # will store indices in the original order

    while pq:
        neg_bound, curr_value, curr_weight, index, taken_sorted = heapq.heappop(pq)
        bound = -neg_bound  # convert back to positive

        # If this branch cannot beat the current best, prune it.
        if bound <= best_value:
            continue

        # If all items have been processed, update best if feasible.
        if index >= n:
            if curr_value > best_value:
                best_value = curr_value
                best_taken_sorted = taken_sorted
            continue

        # Consider the next item in sorted order.
        item_original = sorted_indices[index]
        item_weight = sorted_weights[index]
        item_value = sorted_values[index]

        # Branch 1: Include the item (if it fits).
        if curr_weight + item_weight <= capacity:
            new_value = curr_value + item_value
            new_weight = curr_weight + item_weight
            new_taken_sorted = taken_sorted + [item_original]
            new_bound = compute_bound(index + 1, new_weight, new_value)
            if new_value > best_value:
                best_value = new_value
                best_taken_sorted = new_taken_sorted
            if new_bound > best_value:
                heapq.heappush(pq, (-new_bound, new_value, new_weight, index + 1, new_taken_sorted))

        # Branch 2: Exclude the item.
        exclude_bound = compute_bound(index + 1, curr_weight, curr_value)
        if exclude_bound > best_value:
            heapq.heappush(pq, (-exclude_bound, curr_value, curr_weight, index + 1, taken_sorted))

    # Reconstruct the solution in the original order.
    best_solution = [0] * n
    for i in best_taken_sorted:
        best_solution[i] = 1

    return best_value, best_solution

def main():
    # File path to the instance.
    file_path = "/content/knapsack-01-instances-main/pisinger_instances_01_KP/low-dimensional/f10_l-d_kp_20_879"

    try:
        values, weights, capacity = load_knapsack_instance(file_path)
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found. Please check the path.")
        return
    except Exception as e:
        print(f"Error loading instance: {e}")
        return

    n = len(values)
    print("Loaded knapsack instance:")
    print(f"  Number of items: {n}")
    print(f"  Capacity: {capacity}\n")

    print("Items:")
    print("Index\tValue\tWeight")
    for i in range(n):
        print(f"{i}\t{values[i]}\t{weights[i]}")
    print("-" * 30)

    # Solve using the turbo-boosted solver.
    optimal_value, selected_items = turbo_boosted_knapsack_solver(values, weights, capacity)

    # Compute total weight and value from the selected items to double-check.
    selected_indices = [i for i, bit in enumerate(selected_items) if bit == 1]
    total_weight = sum(weights[i] for i in selected_indices)
    total_value = sum(values[i] for i in selected_indices)

    print("\nOptimal Solution:")
    print(f"  Optimal Value (from solver): {optimal_value}")
    print(f"  Selected Items (binary vector): {selected_items}")
    print(f"  Selected Item Indices: {selected_indices}")
    print(f"  Total Weight of Selected Items: {total_weight}")
    print(f"  Total Value of Selected Items: {total_value}")

    if total_value != optimal_value:
        print("WARNING: The computed total value does not match the solver's optimal value!")

    print("\nDetailed Items Selection:")
    print("Index\tValue\tWeight\tSelected")
    for i in range(n):
        selected_flag = "Yes" if selected_items[i] == 1 else "No"
        print(f"{i}\t{values[i]}\t{weights[i]}\t{selected_flag}")

if __name__ == "__main__":
    main()
```

```python
import heapq
import numpy as np

def load_knapsack_instance(file_path):
    """
    Loads a knapsack instance in the Pisinger format.

    Expected file format:
      - The first line contains two integers: n (number of items) and capacity.
      - The next n lines each contain two integers: value and weight for each item.

    Returns:
      values (list of int): The list of item values.
      weights (list of int): The list of item weights.
      capacity (int): The knapsack capacity.
    """
    with open(file_path, "r") as f:
        lines = f.readlines()

    if not lines:
        raise ValueError("The file is empty.")

    try:
        n, capacity = map(int, lines[0].split())
    except Exception as e:
        raise ValueError("The first line must contain two integers: number of items and capacity.") from e

    if len(lines) < n + 1:
        raise ValueError(f"Expected {n} item lines, but only found {len(lines)-1}.")

    values, weights = [], []
    for i in range(1, n + 1):
        try:
            v, w = map(int, lines[i].split())
        except Exception as e:
            raise ValueError(f"Line {i+1} is not in the expected 'value weight' format.") from e
        values.append(v)
        weights.append(w)
    return values, weights, capacity

def turbo_boosted_knapsack_solver(values, weights, capacity):
    """
    Turbo-boosted Branch and Bound solver for the 0-1 Knapsack Problem.

    Enhancements:
      - Items are sorted by value-to-weight ratio (descending).
      - Precomputed cumulative sums allow a fast (binary search based)
        calculation of a fractional (greedy) upper bound.
      - A best-first search using a priority queue (max-heap) is employed.

    Parameters:
      values: list of item values.
      weights: list of item weights.
      capacity: maximum weight allowed.

    Returns:
      best_value: optimal total value.
      best_solution: binary list (in original order) indicating selected items.
    """
    n = len(values)

    # Sort items by descending ratio (value/weight)
    sorted_indices = sorted(range(n), key=lambda i: values[i] / weights[i], reverse=True)
    sorted_values = [values[i] for i in sorted_indices]
    sorted_weights = [weights[i] for i in sorted_indices]

    # Precompute cumulative sums for the sorted order.
    # cumulative_weight[j] = total weight of items 0 to j-1 in the sorted order.
    # cumulative_value[j]  = total value of items 0 to j-1 in the sorted order.
    cumulative_weight = np.zeros(n + 1)
    cumulative_value = np.zeros(n + 1)
    for j in range(n):
        cumulative_weight[j + 1] = cumulative_weight[j] + sorted_weights[j]
        cumulative_value[j + 1] = cumulative_value[j] + sorted_values[j]

    def compute_bound(index, curr_weight, curr_value):
        """
        Compute an upper bound on the total value that can be achieved,
        starting from the current branch (with items [index, n-1] in the sorted order).

        This function uses binary search on the precomputed cumulative arrays.
        """
        remaining = capacity - curr_weight
        if remaining <= 0:
            return curr_value
        # Find the largest k (>= index) such that:
        # cumulative_weight[k] - cumulative_weight[index] <= remaining
        low = index
        high = n
        while low < high:
            mid = (low + high + 1) // 2
            if cumulative_weight[mid] - cumulative_weight[index] <= remaining:
                low = mid
            else:
                high = mid - 1
        bound_value = curr_value + (cumulative_value[low] - cumulative_value[index])
        # Take a fraction of the next item, if any.
        if low < n:
            remaining_fraction = remaining - (cumulative_weight[low] - cumulative_weight[index])
            if sorted_weights[low] > 0:
                fraction = remaining_fraction / sorted_weights[low]
                bound_value += fraction * sorted_values[low]
        return bound_value

    # Priority queue node: (negative_bound, current_value, current_weight, index, taken_items)
    # We use negative_bound because heapq in Python is a min-heap.
    initial_bound = compute_bound(0, 0, 0)
    pq = [(-initial_bound, 0, 0, 0, [])]

    best_value = 0
    best_taken_sorted = []  # will store indices in the original order

    while pq:
        neg_bound, curr_value, curr_weight, index, taken_sorted = heapq.heappop(pq)
        bound = -neg_bound  # convert back to positive

        # If this branch cannot beat the current best, prune it.
        if bound <= best_value:
            continue

        # If all items have been processed, update best if feasible.
        if index >= n:
            if curr_value > best_value:
                best_value = curr_value
                best_taken_sorted = taken_sorted
            continue

        # Consider the next item in sorted order.
        item_original = sorted_indices[index]
        item_weight = sorted_weights[index]
        item_value = sorted_values[index]

        # Branch 1: Include the item (if it fits).
        if curr_weight + item_weight <= capacity:
            new_value = curr_value + item_value
            new_weight = curr_weight + item_weight
            new_taken_sorted = taken_sorted + [item_original]
            new_bound = compute_bound(index + 1, new_weight, new_value)
            if new_value > best_value:
                best_value = new_value
                best_taken_sorted = new_taken_sorted
            if new_bound > best_value:
                heapq.heappush(pq, (-new_bound, new_value, new_weight, index + 1, new_taken_sorted))

        # Branch 2: Exclude the item.
        exclude_bound = compute_bound(index + 1, curr_weight, curr_value)
        if exclude_bound > best_value:
            heapq.heappush(pq, (-exclude_bound, curr_value, curr_weight, index + 1, taken_sorted))

    # Reconstruct the solution in the original order.
    best_solution = [0] * n
    for i in best_taken_sorted:
        best_solution[i] = 1

    return best_value, best_solution

def main():
    # File path to the instance.
    file_path = "/content/knapsack-01-instances-main/pisinger_instances_01_KP/low-dimensional/f2_l-d_kp_20_878"

    try:
        values, weights, capacity = load_knapsack_instance(file_path)
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found. Please check the path.")
        return
    except Exception as e:
        print(f"Error loading instance: {e}")
        return

    n = len(values)
    print("Loaded knapsack instance:")
    print(f"  Number of items: {n}")
    print(f"  Capacity: {capacity}\n")

    print("Items:")
    print("Index\tValue\tWeight")
    for i in range(n):
        print(f"{i}\t{values[i]}\t{weights[i]}")
    print("-" * 30)

    # Solve using the turbo-boosted solver.
    optimal_value, selected_items = turbo_boosted_knapsack_solver(values, weights, capacity)

    # Compute total weight and value from the selected items to double-check.
    selected_indices = [i for i, bit in enumerate(selected_items) if bit == 1]
    total_weight = sum(weights[i] for i in selected_indices)
    total_value = sum(values[i] for i in selected_indices)

    print("\nOptimal Solution:")
    print(f"  Optimal Value (from solver): {optimal_value}")
    print(f"  Selected Items (binary vector): {selected_items}")
    print(f"  Selected Item Indices: {selected_indices}")
    print(f"  Total Weight of Selected Items: {total_weight}")
    print(f"  Total Value of Selected Items: {total_value}")

    if total_value != optimal_value:
        print("WARNING: The computed total value does not match the solver's optimal value!")

    print("\nDetailed Items Selection:")
    print("Index\tValue\tWeight\tSelected")
    for i in range(n):
        selected_flag = "Yes" if selected_items[i] == 1 else "No"
        print(f"{i}\t{values[i]}\t{weights[i]}\t{selected_flag}")

if __name__ == "__main__":
    main()
```

```python
import heapq
import numpy as np

def load_knapsack_instance(file_path):
    """
    Loads a knapsack instance in the Pisinger format.

    Expected file format:
      - The first line contains two integers: n (number of items) and capacity.
      - The next n lines each contain two integers: value and weight for each item.

    Returns:
      values (list of int): The list of item values.
      weights (list of int): The list of item weights.
      capacity (int): The knapsack capacity.
    """
    with open(file_path, "r") as f:
        lines = f.readlines()

    if not lines:
        raise ValueError("The file is empty.")

    try:
        n, capacity = map(int, lines[0].split())
    except Exception as e:
        raise ValueError("The first line must contain two integers: number of items and capacity.") from e

    if len(lines) < n + 1:
        raise ValueError(f"Expected {n} item lines, but only found {len(lines)-1}.")

    values, weights = [], []
    for i in range(1, n + 1):
        try:
            v, w = map(int, lines[i].split())
        except Exception as e:
            raise ValueError(f"Line {i+1} is not in the expected 'value weight' format.") from e
        values.append(v)
        weights.append(w)
    return values, weights, capacity

def turbo_boosted_knapsack_solver(values, weights, capacity):
    """
    Turbo-boosted Branch and Bound solver for the 0-1 Knapsack Problem.

    Enhancements:
      - Items are sorted by value-to-weight ratio (descending).
      - Precomputed cumulative sums allow a fast (binary search based)
        calculation of a fractional (greedy) upper bound.
      - A best-first search using a priority queue (max-heap) is employed.

    Parameters:
      values: list of item values.
      weights: list of item weights.
      capacity: maximum weight allowed.

    Returns:
      best_value: optimal total value.
      best_solution: binary list (in original order) indicating selected items.
    """
    n = len(values)

    # Sort items by descending ratio (value/weight)
    sorted_indices = sorted(range(n), key=lambda i: values[i] / weights[i], reverse=True)
    sorted_values = [values[i] for i in sorted_indices]
    sorted_weights = [weights[i] for i in sorted_indices]

    # Precompute cumulative sums for the sorted order.
    # cumulative_weight[j] = total weight of items 0 to j-1 in the sorted order.
    # cumulative_value[j]  = total value of items 0 to j-1 in the sorted order.
    cumulative_weight = np.zeros(n + 1)
    cumulative_value = np.zeros(n + 1)
    for j in range(n):
        cumulative_weight[j + 1] = cumulative_weight[j] + sorted_weights[j]
        cumulative_value[j + 1] = cumulative_value[j] + sorted_values[j]

    def compute_bound(index, curr_weight, curr_value):
        """
        Compute an upper bound on the total value that can be achieved,
        starting from the current branch (with items [index, n-1] in the sorted order).

        This function uses binary search on the precomputed cumulative arrays.
        """
        remaining = capacity - curr_weight
        if remaining <= 0:
            return curr_value
        # Find the largest k (>= index) such that:
        # cumulative_weight[k] - cumulative_weight[index] <= remaining
        low = index
        high = n
        while low < high:
            mid = (low + high + 1) // 2
            if cumulative_weight[mid] - cumulative_weight[index] <= remaining:
                low = mid
            else:
                high = mid - 1
        bound_value = curr_value + (cumulative_value[low] - cumulative_value[index])
        # Take a fraction of the next item, if any.
        if low < n:
            remaining_fraction = remaining - (cumulative_weight[low] - cumulative_weight[index])
            if sorted_weights[low] > 0:
                fraction = remaining_fraction / sorted_weights[low]
                bound_value += fraction * sorted_values[low]
        return bound_value

    # Priority queue node: (negative_bound, current_value, current_weight, index, taken_items)
    # We use negative_bound because heapq in Python is a min-heap.
    initial_bound = compute_bound(0, 0, 0)
    pq = [(-initial_bound, 0, 0, 0, [])]

    best_value = 0
    best_taken_sorted = []  # will store indices in the original order

    while pq:
        neg_bound, curr_value, curr_weight, index, taken_sorted = heapq.heappop(pq)
        bound = -neg_bound  # convert back to positive

        # If this branch cannot beat the current best, prune it.
        if bound <= best_value:
            continue

        # If all items have been processed, update best if feasible.
        if index >= n:
            if curr_value > best_value:
                best_value = curr_value
                best_taken_sorted = taken_sorted
            continue

        # Consider the next item in sorted order.
        item_original = sorted_indices[index]
        item_weight = sorted_weights[index]
        item_value = sorted_values[index]

        # Branch 1: Include the item (if it fits).
        if curr_weight + item_weight <= capacity:
            new_value = curr_value + item_value
            new_weight = curr_weight + item_weight
            new_taken_sorted = taken_sorted + [item_original]
            new_bound = compute_bound(index + 1, new_weight, new_value)
            if new_value > best_value:
                best_value = new_value
                best_taken_sorted = new_taken_sorted
            if new_bound > best_value:
                heapq.heappush(pq, (-new_bound, new_value, new_weight, index + 1, new_taken_sorted))

        # Branch 2: Exclude the item.
        exclude_bound = compute_bound(index + 1, curr_weight, curr_value)
        if exclude_bound > best_value:
            heapq.heappush(pq, (-exclude_bound, curr_value, curr_weight, index + 1, taken_sorted))

    # Reconstruct the solution in the original order.
    best_solution = [0] * n
    for i in best_taken_sorted:
        best_solution[i] = 1

    return best_value, best_solution

def main():
    # File path to the instance.
    file_path = "/content/knapsack-01-instances-main/pisinger_instances_01_KP/low-dimensional/f4_l-d_kp_4_11"

    try:
        values, weights, capacity = load_knapsack_instance(file_path)
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found. Please check the path.")
        return
    except Exception as e:
        print(f"Error loading instance: {e}")
        return

    n = len(values)
    print("Loaded knapsack instance:")
    print(f"  Number of items: {n}")
    print(f"  Capacity: {capacity}\n")

    print("Items:")
    print("Index\tValue\tWeight")
    for i in range(n):
        print(f"{i}\t{values[i]}\t{weights[i]}")
    print("-" * 30)

    # Solve using the turbo-boosted solver.
    optimal_value, selected_items = turbo_boosted_knapsack_solver(values, weights, capacity)

    # Compute total weight and value from the selected items to double-check.
    selected_indices = [i for i, bit in enumerate(selected_items) if bit == 1]
    total_weight = sum(weights[i] for i in selected_indices)
    total_value = sum(values[i] for i in selected_indices)

    print("\nOptimal Solution:")
    print(f"  Optimal Value (from solver): {optimal_value}")
    print(f"  Selected Items (binary vector): {selected_items}")
    print(f"  Selected Item Indices: {selected_indices}")
    print(f"  Total Weight of Selected Items: {total_weight}")
    print(f"  Total Value of Selected Items: {total_value}")

    if total_value != optimal_value:
        print("WARNING: The computed total value does not match the solver's optimal value!")

    print("\nDetailed Items Selection:")
    print("Index\tValue\tWeight\tSelected")
    for i in range(n):
        selected_flag = "Yes" if selected_items[i] == 1 else "No"
        print(f"{i}\t{values[i]}\t{weights[i]}\t{selected_flag}")

if __name__ == "__main__":
    main()
```

```python
import heapq
import numpy as np

def load_knapsack_instance(file_path):
    """
    Loads a knapsack instance in the Pisinger format.

    Expected file format:
      - The first line contains two integers: n (number of items) and capacity.
      - The next n lines each contain two integers: value and weight for each item.

    Returns:
      values (list of int): The list of item values.
      weights (list of int): The list of item weights.
      capacity (int): The knapsack capacity.
    """
    with open(file_path, "r") as f:
        lines = f.readlines()

    if not lines:
        raise ValueError("The file is empty.")

    try:
        n, capacity = map(int, lines[0].split())
    except Exception as e:
        raise ValueError("The first line must contain two integers: number of items and capacity.") from e

    if len(lines) < n + 1:
        raise ValueError(f"Expected {n} item lines, but only found {len(lines)-1}.")

    values, weights = [], []
    for i in range(1, n + 1):
        try:
            v, w = map(int, lines[i].split())
        except Exception as e:
            raise ValueError(f"Line {i+1} is not in the expected 'value weight' format.") from e
        values.append(v)
        weights.append(w)
    return values, weights, capacity

def turbo_boosted_knapsack_solver(values, weights, capacity):
    """
    Turbo-boosted Branch and Bound solver for the 0-1 Knapsack Problem.

    Enhancements:
      - Items are sorted by value-to-weight ratio (descending).
      - Precomputed cumulative sums allow a fast (binary search based)
        calculation of a fractional (greedy) upper bound.
      - A best-first search using a priority queue (max-heap) is employed.

    Parameters:
      values: list of item values.
      weights: list of item weights.
      capacity: maximum weight allowed.

    Returns:
      best_value: optimal total value.
      best_solution: binary list (in original order) indicating selected items.
    """
    n = len(values)

    # Sort items by descending ratio (value/weight)
    sorted_indices = sorted(range(n), key=lambda i: values[i] / weights[i], reverse=True)
    sorted_values = [values[i] for i in sorted_indices]
    sorted_weights = [weights[i] for i in sorted_indices]

    # Precompute cumulative sums for the sorted order.
    # cumulative_weight[j] = total weight of items 0 to j-1 in the sorted order.
    # cumulative_value[j]  = total value of items 0 to j-1 in the sorted order.
    cumulative_weight = np.zeros(n + 1)
    cumulative_value = np.zeros(n + 1)
    for j in range(n):
        cumulative_weight[j + 1] = cumulative_weight[j] + sorted_weights[j]
        cumulative_value[j + 1] = cumulative_value[j] + sorted_values[j]

    def compute_bound(index, curr_weight, curr_value):
        """
        Compute an upper bound on the total value that can be achieved,
        starting from the current branch (with items [index, n-1] in the sorted order).

        This function uses binary search on the precomputed cumulative arrays.
        """
        remaining = capacity - curr_weight
        if remaining <= 0:
            return curr_value
        # Find the largest k (>= index) such that:
        # cumulative_weight[k] - cumulative_weight[index] <= remaining
        low = index
        high = n
        while low < high:
            mid = (low + high + 1) // 2
            if cumulative_weight[mid] - cumulative_weight[index] <= remaining:
                low = mid
            else:
                high = mid - 1
        bound_value = curr_value + (cumulative_value[low] - cumulative_value[index])
        # Take a fraction of the next item, if any.
        if low < n:
            remaining_fraction = remaining - (cumulative_weight[low] - cumulative_weight[index])
            if sorted_weights[low] > 0:
                fraction = remaining_fraction / sorted_weights[low]
                bound_value += fraction * sorted_values[low]
        return bound_value

    # Priority queue node: (negative_bound, current_value, current_weight, index, taken_items)
    # We use negative_bound because heapq in Python is a min-heap.
    initial_bound = compute_bound(0, 0, 0)
    pq = [(-initial_bound, 0, 0, 0, [])]

    best_value = 0
    best_taken_sorted = []  # will store indices in the original order

    while pq:
        neg_bound, curr_value, curr_weight, index, taken_sorted = heapq.heappop(pq)
        bound = -neg_bound  # convert back to positive

        # If this branch cannot beat the current best, prune it.
        if bound <= best_value:
            continue

        # If all items have been processed, update best if feasible.
        if index >= n:
            if curr_value > best_value:
                best_value = curr_value
                best_taken_sorted = taken_sorted
            continue

        # Consider the next item in sorted order.
        item_original = sorted_indices[index]
        item_weight = sorted_weights[index]
        item_value = sorted_values[index]

        # Branch 1: Include the item (if it fits).
        if curr_weight + item_weight <= capacity:
            new_value = curr_value + item_value
            new_weight = curr_weight + item_weight
            new_taken_sorted = taken_sorted + [item_original]
            new_bound = compute_bound(index + 1, new_weight, new_value)
            if new_value > best_value:
                best_value = new_value
                best_taken_sorted = new_taken_sorted
            if new_bound > best_value:
                heapq.heappush(pq, (-new_bound, new_value, new_weight, index + 1, new_taken_sorted))

        # Branch 2: Exclude the item.
        exclude_bound = compute_bound(index + 1, curr_weight, curr_value)
        if exclude_bound > best_value:
            heapq.heappush(pq, (-exclude_bound, curr_value, curr_weight, index + 1, taken_sorted))

    # Reconstruct the solution in the original order.
    best_solution = [0] * n
    for i in best_taken_sorted:
        best_solution[i] = 1

    return best_value, best_solution

def main():
    # File path to the instance.
    file_path = "/content/knapsack-01-instances-main/pisinger_instances_01_KP/low-dimensional/f6_l-d_kp_10_60"

    try:
        values, weights, capacity = load_knapsack_instance(file_path)
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found. Please check the path.")
        return
    except Exception as e:
        print(f"Error loading instance: {e}")
        return

    n = len(values)
    print("Loaded knapsack instance:")
    print(f"  Number of items: {n}")
    print(f"  Capacity: {capacity}\n")

    print("Items:")
    print("Index\tValue\tWeight")
    for i in range(n):
        print(f"{i}\t{values[i]}\t{weights[i]}")
    print("-" * 30)

    # Solve using the turbo-boosted solver.
    optimal_value, selected_items = turbo_boosted_knapsack_solver(values, weights, capacity)

    # Compute total weight and value from the selected items to double-check.
    selected_indices = [i for i, bit in enumerate(selected_items) if bit == 1]
    total_weight = sum(weights[i] for i in selected_indices)
    total_value = sum(values[i] for i in selected_indices)

    print("\nOptimal Solution:")
    print(f"  Optimal Value (from solver): {optimal_value}")
    print(f"  Selected Items (binary vector): {selected_items}")
    print(f"  Selected Item Indices: {selected_indices}")
    print(f"  Total Weight of Selected Items: {total_weight}")
    print(f"  Total Value of Selected Items: {total_value}")

    if total_value != optimal_value:
        print("WARNING: The computed total value does not match the solver's optimal value!")

    print("\nDetailed Items Selection:")
    print("Index\tValue\tWeight\tSelected")
    for i in range(n):
        selected_flag = "Yes" if selected_items[i] == 1 else "No"
        print(f"{i}\t{values[i]}\t{weights[i]}\t{selected_flag}")

if __name__ == "__main__":
    main()
```

```python
import heapq
import numpy as np

def load_knapsack_instance(file_path):
    """
    Loads a knapsack instance in the Pisinger format.

    Expected file format:
      - The first line contains two integers: n (number of items) and capacity.
      - The next n lines each contain two integers: value and weight for each item.

    Returns:
      values (list of int): The list of item values.
      weights (list of int): The list of item weights.
      capacity (int): The knapsack capacity.
    """
    with open(file_path, "r") as f:
        lines = f.readlines()

    if not lines:
        raise ValueError("The file is empty.")

    try:
        n, capacity = map(int, lines[0].split())
    except Exception as e:
        raise ValueError("The first line must contain two integers: number of items and capacity.") from e

    if len(lines) < n + 1:
        raise ValueError(f"Expected {n} item lines, but only found {len(lines)-1}.")

    values, weights = [], []
    for i in range(1, n + 1):
        try:
            v, w = map(int, lines[i].split())
        except Exception as e:
            raise ValueError(f"Line {i+1} is not in the expected 'value weight' format.") from e
        values.append(v)
        weights.append(w)
    return values, weights, capacity

def turbo_boosted_knapsack_solver(values, weights, capacity):
    """
    Turbo-boosted Branch and Bound solver for the 0-1 Knapsack Problem.

    Enhancements:
      - Items are sorted by value-to-weight ratio (descending).
      - Precomputed cumulative sums allow a fast (binary search based)
        calculation of a fractional (greedy) upper bound.
      - A best-first search using a priority queue (max-heap) is employed.

    Parameters:
      values: list of item values.
      weights: list of item weights.
      capacity: maximum weight allowed.

    Returns:
      best_value: optimal total value.
      best_solution: binary list (in original order) indicating selected items.
    """
    n = len(values)

    # Sort items by descending ratio (value/weight)
    sorted_indices = sorted(range(n), key=lambda i: values[i] / weights[i], reverse=True)
    sorted_values = [values[i] for i in sorted_indices]
    sorted_weights = [weights[i] for i in sorted_indices]

    # Precompute cumulative sums for the sorted order.
    # cumulative_weight[j] = total weight of items 0 to j-1 in the sorted order.
    # cumulative_value[j]  = total value of items 0 to j-1 in the sorted order.
    cumulative_weight = np.zeros(n + 1)
    cumulative_value = np.zeros(n + 1)
    for j in range(n):
        cumulative_weight[j + 1] = cumulative_weight[j] + sorted_weights[j]
        cumulative_value[j + 1] = cumulative_value[j] + sorted_values[j]

    def compute_bound(index, curr_weight, curr_value):
        """
        Compute an upper bound on the total value that can be achieved,
        starting from the current branch (with items [index, n-1] in the sorted order).

        This function uses binary search on the precomputed cumulative arrays.
        """
        remaining = capacity - curr_weight
        if remaining <= 0:
            return curr_value
        # Find the largest k (>= index) such that:
        # cumulative_weight[k] - cumulative_weight[index] <= remaining
        low = index
        high = n
        while low < high:
            mid = (low + high + 1) // 2
            if cumulative_weight[mid] - cumulative_weight[index] <= remaining:
                low = mid
            else:
                high = mid - 1
        bound_value = curr_value + (cumulative_value[low] - cumulative_value[index])
        # Take a fraction of the next item, if any.
        if low < n:
            remaining_fraction = remaining - (cumulative_weight[low] - cumulative_weight[index])
            if sorted_weights[low] > 0:
                fraction = remaining_fraction / sorted_weights[low]
                bound_value += fraction * sorted_values[low]
        return bound_value

    # Priority queue node: (negative_bound, current_value, current_weight, index, taken_items)
    # We use negative_bound because heapq in Python is a min-heap.
    initial_bound = compute_bound(0, 0, 0)
    pq = [(-initial_bound, 0, 0, 0, [])]

    best_value = 0
    best_taken_sorted = []  # will store indices in the original order

    while pq:
        neg_bound, curr_value, curr_weight, index, taken_sorted = heapq.heappop(pq)
        bound = -neg_bound  # convert back to positive

        # If this branch cannot beat the current best, prune it.
        if bound <= best_value:
            continue

        # If all items have been processed, update best if feasible.
        if index >= n:
            if curr_value > best_value:
                best_value = curr_value
                best_taken_sorted = taken_sorted
            continue

        # Consider the next item in sorted order.
        item_original = sorted_indices[index]
        item_weight = sorted_weights[index]
        item_value = sorted_values[index]

        # Branch 1: Include the item (if it fits).
        if curr_weight + item_weight <= capacity:
            new_value = curr_value + item_value
            new_weight = curr_weight + item_weight
            new_taken_sorted = taken_sorted + [item_original]
            new_bound = compute_bound(index + 1, new_weight, new_value)
            if new_value > best_value:
                best_value = new_value
                best_taken_sorted = new_taken_sorted
            if new_bound > best_value:
                heapq.heappush(pq, (-new_bound, new_value, new_weight, index + 1, new_taken_sorted))

        # Branch 2: Exclude the item.
        exclude_bound = compute_bound(index + 1, curr_weight, curr_value)
        if exclude_bound > best_value:
            heapq.heappush(pq, (-exclude_bound, curr_value, curr_weight, index + 1, taken_sorted))

    # Reconstruct the solution in the original order.
    best_solution = [0] * n
    for i in best_taken_sorted:
        best_solution[i] = 1

    return best_value, best_solution

def main():
    # File path to the instance.
    file_path = "/content/knapsack-01-instances-main/pisinger_instances_01_KP/low-dimensional/f8_l-d_kp_23_10000"

    try:
        values, weights, capacity = load_knapsack_instance(file_path)
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found. Please check the path.")
        return
    except Exception as e:
        print(f"Error loading instance: {e}")
        return

    n = len(values)
    print("Loaded knapsack instance:")
    print(f"  Number of items: {n}")
    print(f"  Capacity: {capacity}\n")

    print("Items:")
    print("Index\tValue\tWeight")
    for i in range(n):
        print(f"{i}\t{values[i]}\t{weights[i]}")
    print("-" * 30)

    # Solve using the turbo-boosted solver.
    optimal_value, selected_items = turbo_boosted_knapsack_solver(values, weights, capacity)

    # Compute total weight and value from the selected items to double-check.
    selected_indices = [i for i, bit in enumerate(selected_items) if bit == 1]
    total_weight = sum(weights[i] for i in selected_indices)
    total_value = sum(values[i] for i in selected_indices)

    print("\nOptimal Solution:")
    print(f"  Optimal Value (from solver): {optimal_value}")
    print(f"  Selected Items (binary vector): {selected_items}")
    print(f"  Selected Item Indices: {selected_indices}")
    print(f"  Total Weight of Selected Items: {total_weight}")
    print(f"  Total Value of Selected Items: {total_value}")

    if total_value != optimal_value:
        print("WARNING: The computed total value does not match the solver's optimal value!")

    print("\nDetailed Items Selection:")
    print("Index\tValue\tWeight\tSelected")
    for i in range(n):
        selected_flag = "Yes" if selected_items[i] == 1 else "No"
        print(f"{i}\t{values[i]}\t{weights[i]}\t{selected_flag}")

if __name__ == "__main__":
    main()
```

```python
import heapq
import numpy as np
import sys

class Tee:
    """
    A helper class that duplicates output to multiple streams.
    """
    def __init__(self, *files):
        self.files = files

    def write(self, data):
        for f in self.files:
            f.write(data)

    def flush(self):
        for f in self.files:
            f.flush()

def load_knapsack_instance(file_path):
    """
    Loads a knapsack instance in the Pisinger format.

    Expected file format:
      - The first line contains two integers: n (number of items) and capacity.
      - The next n lines each contain two integers: value and weight for each item.

    Returns:
      values (list of int): The list of item values.
      weights (list of int): The list of item weights.
      capacity (int): The knapsack capacity.
    """
    with open(file_path, "r") as f:
        lines = f.readlines()

    if not lines:
        raise ValueError("The file is empty.")

    try:
        n, capacity = map(int, lines[0].split())
    except Exception as e:
        raise ValueError("The first line must contain two integers: number of items and capacity.") from e

    if len(lines) < n + 1:
        raise ValueError(f"Expected {n} item lines, but only found {len(lines)-1}.")

    values, weights = [], []
    for i in range(1, n + 1):
        try:
            v, w = map(int, lines[i].split())
        except Exception as e:
            raise ValueError(f"Line {i+1} is not in the expected 'value weight' format.") from e
        values.append(v)
        weights.append(w)
    return values, weights, capacity

def turbo_boosted_knapsack_solver(values, weights, capacity):
    """
    Turbo-boosted Branch and Bound solver for the 0-1 Knapsack Problem.

    Enhancements:
      - Items are sorted by value-to-weight ratio (descending).
      - Precomputed cumulative sums allow a fast (binary search based)
        calculation of a fractional (greedy) upper bound.
      - A best-first search using a priority queue (max-heap) is employed.

    Parameters:
      values: list of item values.
      weights: list of item weights.
      capacity: maximum weight allowed.

    Returns:
      best_value: optimal total value.
      best_solution: binary list (in original order) indicating selected items.
    """
    n = len(values)

    # Sort items by descending ratio (value/weight)
    sorted_indices = sorted(range(n), key=lambda i: values[i] / weights[i], reverse=True)
    sorted_values = [values[i] for i in sorted_indices]
    sorted_weights = [weights[i] for i in sorted_indices]

    # Precompute cumulative sums for the sorted order.
    cumulative_weight = np.zeros(n + 1)
    cumulative_value = np.zeros(n + 1)
    for j in range(n):
        cumulative_weight[j + 1] = cumulative_weight[j] + sorted_weights[j]
        cumulative_value[j + 1] = cumulative_value[j] + sorted_values[j]

    def compute_bound(index, curr_weight, curr_value):
        """
        Compute an upper bound on the total value that can be achieved,
        starting from the current branch (with items [index, n-1] in the sorted order).

        Uses binary search on the precomputed cumulative arrays.
        """
        remaining = capacity - curr_weight
        if remaining <= 0:
            return curr_value
        low = index
        high = n
        while low < high:
            mid = (low + high + 1) // 2
            if cumulative_weight[mid] - cumulative_weight[index] <= remaining:
                low = mid
            else:
                high = mid - 1
        bound_value = curr_value + (cumulative_value[low] - cumulative_value[index])
        if low < n:
            remaining_fraction = remaining - (cumulative_weight[low] - cumulative_weight[index])
            if sorted_weights[low] > 0:
                fraction = remaining_fraction / sorted_weights[low]
                bound_value += fraction * sorted_values[low]
        return bound_value

    # Priority queue node: (negative_bound, current_value, current_weight, index, taken_items)
    initial_bound = compute_bound(0, 0, 0)
    pq = [(-initial_bound, 0, 0, 0, [])]

    best_value = 0
    best_taken_sorted = []  # will store indices in original order

    while pq:
        neg_bound, curr_value, curr_weight, index, taken_sorted = heapq.heappop(pq)
        bound = -neg_bound  # convert back to positive

        if bound <= best_value:
            continue

        if index >= n:
            if curr_value > best_value:
                best_value = curr_value
                best_taken_sorted = taken_sorted
            continue

        # Consider the next item in sorted order.
        item_original = sorted_indices[index]
        item_weight = sorted_weights[index]
        item_value = sorted_values[index]

        # Branch 1: Include the item (if it fits).
        if curr_weight + item_weight <= capacity:
            new_value = curr_value + item_value
            new_weight = curr_weight + item_weight
            new_taken_sorted = taken_sorted + [item_original]
            new_bound = compute_bound(index + 1, new_weight, new_value)
            if new_value > best_value:
                best_value = new_value
                best_taken_sorted = new_taken_sorted
            if new_bound > best_value:
                heapq.heappush(pq, (-new_bound, new_value, new_weight, index + 1, new_taken_sorted))

        # Branch 2: Exclude the item.
        exclude_bound = compute_bound(index + 1, curr_weight, curr_value)
        if exclude_bound > best_value:
            heapq.heappush(pq, (-exclude_bound, curr_value, curr_weight, index + 1, taken_sorted))

    # Reconstruct the solution in the original order.
    best_solution = [0] * n
    for i in best_taken_sorted:
        best_solution[i] = 1

    return best_value, best_solution

def print_table(header, rows, max_lines=20):
    """
    Utility function to print a table in a summarized form if there are many rows.
    If total rows > max_lines, prints first half and last half with an ellipsis in between.
    """
    print(header)
    n = len(rows)
    if n <= max_lines:
        for row in rows:
            print("\t".join(map(str, row)))
    else:
        half = max_lines // 2
        for row in rows[:half]:
            print("\t".join(map(str, row)))
        print("...")
        for row in rows[-half:]:
            print("\t".join(map(str, row)))

def main():
    # File path to the knapsack instance.
    file_path = "/content/knapsack-01-instances-main/pisinger_instances_01_KP/large_scale/knapPI_1_10000_1000_1"

    try:
        values, weights, capacity = load_knapsack_instance(file_path)
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found. Please check the path.")
        return
    except Exception as e:
        print(f"Error loading instance: {e}")
        return

    n = len(values)
    print("Loaded knapsack instance:")
    print(f"  Number of items: {n}")
    print(f"  Capacity: {capacity}\n")

    # Prepare and print the Items table.
    items_rows = [["Index", "Value", "Weight"]]
    for i in range(n):
        items_rows.append([i, values[i], weights[i]])

    print("Items:")
    MAX_LINES_TO_PRINT = 20  # Adjust as needed.
    print_table("Index\tValue\tWeight", items_rows[1:], max_lines=MAX_LINES_TO_PRINT)
    print("-" * 30)

    # Solve using the turbo-boosted solver.
    optimal_value, selected_items = turbo_boosted_knapsack_solver(values, weights, capacity)

    # Double-check: recompute total weight and total value from selected items.
    selected_indices = [i for i, bit in enumerate(selected_items) if bit == 1]
    total_weight = sum(weights[i] for i in selected_indices)
    total_value = sum(values[i] for i in selected_indices)

    print("\nOptimal Solution:")
    print(f"  Optimal Value (from solver): {optimal_value}")
    print(f"  Selected Items (binary vector): {selected_items}")
    print(f"  Selected Item Indices: {selected_indices}")
    print(f"  Total Weight of Selected Items: {total_weight}")
    print(f"  Total Value of Selected Items: {total_value}")

    if total_value != optimal_value:
        print("WARNING: The computed total value does not match the solver's optimal value!")

    # Prepare and print the Detailed Items Selection table.
    detailed_rows = [["Index", "Value", "Weight", "Selected"]]
    for i in range(n):
        selected_flag = "Yes" if selected_items[i] == 1 else "No"
        detailed_rows.append([i, values[i], weights[i], selected_flag])

    print("\nDetailed Items Selection:")
    print_table("Index\tValue\tWeight\tSelected", detailed_rows[1:], max_lines=MAX_LINES_TO_PRINT)

    # Print the final optimal solution summary at the end.
    print("\n" + "=" * 40)
    print("FINAL OPTIMAL SOLUTION SUMMARY")
    print("=" * 40)
    print(f"Optimal Value: {optimal_value}")
    print(f"Selected Items (binary vector): {selected_items}")
    print(f"Selected Item Indices: {selected_indices}")
    print(f"Total Weight: {total_weight}")
    print(f"Total Value: {total_value}")
    print("=" * 40)

if __name__ == "__main__":
    # Open output file and redirect sys.stdout using a try/finally block to ensure restoration.
    out_file = open("knapsack_output.txt", "w")
    original_stdout = sys.stdout
    sys.stdout = Tee(original_stdout, out_file)
    try:
        main()
    finally:
        # Flush and restore the original stdout, then close the file.
        sys.stdout.flush()
        sys.stdout = original_stdout
        out_file.close()
```

```python
import heapq
import numpy as np
import sys

class Tee:
    """
    A helper class that duplicates output to multiple streams.
    """
    def __init__(self, *files):
        self.files = files

    def write(self, data):
        for f in self.files:
            f.write(data)

    def flush(self):
        for f in self.files:
            f.flush()

def load_knapsack_instance(file_path):
    """
    Loads a knapsack instance in the Pisinger format.

    Expected file format:
      - The first line contains two integers: n (number of items) and capacity.
      - The next n lines each contain two integers: value and weight for each item.

    Returns:
      values (list of int): The list of item values.
      weights (list of int): The list of item weights.
      capacity (int): The knapsack capacity.
    """
    with open(file_path, "r") as f:
        lines = f.readlines()

    if not lines:
        raise ValueError("The file is empty.")

    try:
        n, capacity = map(int, lines[0].split())
    except Exception as e:
        raise ValueError("The first line must contain two integers: number of items and capacity.") from e

    if len(lines) < n + 1:
        raise ValueError(f"Expected {n} item lines, but only found {len(lines)-1}.")

    values, weights = [], []
    for i in range(1, n + 1):
        try:
            v, w = map(int, lines[i].split())
        except Exception as e:
            raise ValueError(f"Line {i+1} is not in the expected 'value weight' format.") from e
        values.append(v)
        weights.append(w)
    return values, weights, capacity

def turbo_boosted_knapsack_solver(values, weights, capacity):
    """
    Turbo-boosted Branch and Bound solver for the 0-1 Knapsack Problem.

    Enhancements:
      - Items are sorted by value-to-weight ratio (descending).
      - Precomputed cumulative sums allow a fast (binary search based)
        calculation of a fractional (greedy) upper bound.
      - A best-first search using a priority queue (max-heap) is employed.

    Parameters:
      values: list of item values.
      weights: list of item weights.
      capacity: maximum weight allowed.

    Returns:
      best_value: optimal total value.
      best_solution: binary list (in original order) indicating selected items.
    """
    n = len(values)

    # Sort items by descending ratio (value/weight)
    sorted_indices = sorted(range(n), key=lambda i: values[i] / weights[i], reverse=True)
    sorted_values = [values[i] for i in sorted_indices]
    sorted_weights = [weights[i] for i in sorted_indices]

    # Precompute cumulative sums for the sorted order.
    cumulative_weight = np.zeros(n + 1)
    cumulative_value = np.zeros(n + 1)
    for j in range(n):
        cumulative_weight[j + 1] = cumulative_weight[j] + sorted_weights[j]
        cumulative_value[j + 1] = cumulative_value[j] + sorted_values[j]

    def compute_bound(index, curr_weight, curr_value):
        """
        Compute an upper bound on the total value that can be achieved,
        starting from the current branch (with items [index, n-1] in the sorted order).

        Uses binary search on the precomputed cumulative arrays.
        """
        remaining = capacity - curr_weight
        if remaining <= 0:
            return curr_value
        low = index
        high = n
        while low < high:
            mid = (low + high + 1) // 2
            if cumulative_weight[mid] - cumulative_weight[index] <= remaining:
                low = mid
            else:
                high = mid - 1
        bound_value = curr_value + (cumulative_value[low] - cumulative_value[index])
        if low < n:
            remaining_fraction = remaining - (cumulative_weight[low] - cumulative_weight[index])
            if sorted_weights[low] > 0:
                fraction = remaining_fraction / sorted_weights[low]
                bound_value += fraction * sorted_values[low]
        return bound_value

    # Priority queue node: (negative_bound, current_value, current_weight, index, taken_items)
    initial_bound = compute_bound(0, 0, 0)
    pq = [(-initial_bound, 0, 0, 0, [])]

    best_value = 0
    best_taken_sorted = []  # will store indices in original order

    while pq:
        neg_bound, curr_value, curr_weight, index, taken_sorted = heapq.heappop(pq)
        bound = -neg_bound  # convert back to positive

        if bound <= best_value:
            continue

        if index >= n:
            if curr_value > best_value:
                best_value = curr_value
                best_taken_sorted = taken_sorted
            continue

        # Consider the next item in sorted order.
        item_original = sorted_indices[index]
        item_weight = sorted_weights[index]
        item_value = sorted_values[index]

        # Branch 1: Include the item (if it fits).
        if curr_weight + item_weight <= capacity:
            new_value = curr_value + item_value
            new_weight = curr_weight + item_weight
            new_taken_sorted = taken_sorted + [item_original]
            new_bound = compute_bound(index + 1, new_weight, new_value)
            if new_value > best_value:
                best_value = new_value
                best_taken_sorted = new_taken_sorted
            if new_bound > best_value:
                heapq.heappush(pq, (-new_bound, new_value, new_weight, index + 1, new_taken_sorted))

        # Branch 2: Exclude the item.
        exclude_bound = compute_bound(index + 1, curr_weight, curr_value)
        if exclude_bound > best_value:
            heapq.heappush(pq, (-exclude_bound, curr_value, curr_weight, index + 1, taken_sorted))

    # Reconstruct the solution in the original order.
    best_solution = [0] * n
    for i in best_taken_sorted:
        best_solution[i] = 1

    return best_value, best_solution

def print_table(header, rows, max_lines=20):
    """
    Utility function to print a table in a summarized form if there are many rows.
    If total rows > max_lines, prints first half and last half with an ellipsis in between.
    """
    print(header)
    n = len(rows)
    if n <= max_lines:
        for row in rows:
            print("\t".join(map(str, row)))
    else:
        half = max_lines // 2
        for row in rows[:half]:
            print("\t".join(map(str, row)))
        print("...")
        for row in rows[-half:]:
            print("\t".join(map(str, row)))

def main():
    # File path to the knapsack instance.
    file_path = "/content/knapsack-01-instances-main/pisinger_instances_01_KP/large_scale/knapPI_1_100_1000_1"

    try:
        values, weights, capacity = load_knapsack_instance(file_path)
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found. Please check the path.")
        return
    except Exception as e:
        print(f"Error loading instance: {e}")
        return

    n = len(values)
    print("Loaded knapsack instance:")
    print(f"  Number of items: {n}")
    print(f"  Capacity: {capacity}\n")

    # Prepare and print the Items table.
    items_rows = [["Index", "Value", "Weight"]]
    for i in range(n):
        items_rows.append([i, values[i], weights[i]])

    print("Items:")
    MAX_LINES_TO_PRINT = 20  # Adjust as needed.
    print_table("Index\tValue\tWeight", items_rows[1:], max_lines=MAX_LINES_TO_PRINT)
    print("-" * 30)

    # Solve using the turbo-boosted solver.
    optimal_value, selected_items = turbo_boosted_knapsack_solver(values, weights, capacity)

    # Double-check: recompute total weight and total value from selected items.
    selected_indices = [i for i, bit in enumerate(selected_items) if bit == 1]
    total_weight = sum(weights[i] for i in selected_indices)
    total_value = sum(values[i] for i in selected_indices)

    print("\nOptimal Solution:")
    print(f"  Optimal Value (from solver): {optimal_value}")
    print(f"  Selected Items (binary vector): {selected_items}")
    print(f"  Selected Item Indices: {selected_indices}")
    print(f"  Total Weight of Selected Items: {total_weight}")
    print(f"  Total Value of Selected Items: {total_value}")

    if total_value != optimal_value:
        print("WARNING: The computed total value does not match the solver's optimal value!")

    # Prepare and print the Detailed Items Selection table.
    detailed_rows = [["Index", "Value", "Weight", "Selected"]]
    for i in range(n):
        selected_flag = "Yes" if selected_items[i] == 1 else "No"
        detailed_rows.append([i, values[i], weights[i], selected_flag])

    print("\nDetailed Items Selection:")
    print_table("Index\tValue\tWeight\tSelected", detailed_rows[1:], max_lines=MAX_LINES_TO_PRINT)

    # Print the final optimal solution summary at the end.
    print("\n" + "=" * 40)
    print("FINAL OPTIMAL SOLUTION SUMMARY")
    print("=" * 40)
    print(f"Optimal Value: {optimal_value}")
    print(f"Selected Items (binary vector): {selected_items}")
    print(f"Selected Item Indices: {selected_indices}")
    print(f"Total Weight: {total_weight}")
    print(f"Total Value: {total_value}")
    print("=" * 40)

if __name__ == "__main__":
    # Open output file and redirect sys.stdout using a try/finally block to ensure restoration.
    out_file = open("knapsack_output.txt", "w")
    original_stdout = sys.stdout
    sys.stdout = Tee(original_stdout, out_file)
    try:
        main()
    finally:
        # Flush and restore the original stdout, then close the file.
        sys.stdout.flush()
        sys.stdout = original_stdout
        out_file.close()
```

```python
import heapq
import numpy as np
import sys

class Tee:
    """
    A helper class that duplicates output to multiple streams.
    """
    def __init__(self, *files):
        self.files = files

    def write(self, data):
        for f in self.files:
            f.write(data)

    def flush(self):
        for f in self.files:
            f.flush()

def load_knapsack_instance(file_path):
    """
    Loads a knapsack instance in the Pisinger format.

    Expected file format:
      - The first line contains two integers: n (number of items) and capacity.
      - The next n lines each contain two integers: value and weight for each item.

    Returns:
      values (list of int): The list of item values.
      weights (list of int): The list of item weights.
      capacity (int): The knapsack capacity.
    """
    with open(file_path, "r") as f:
        lines = f.readlines()

    if not lines:
        raise ValueError("The file is empty.")

    try:
        n, capacity = map(int, lines[0].split())
    except Exception as e:
        raise ValueError("The first line must contain two integers: number of items and capacity.") from e

    if len(lines) < n + 1:
        raise ValueError(f"Expected {n} item lines, but only found {len(lines)-1}.")

    values, weights = [], []
    for i in range(1, n + 1):
        try:
            v, w = map(int, lines[i].split())
        except Exception as e:
            raise ValueError(f"Line {i+1} is not in the expected 'value weight' format.") from e
        values.append(v)
        weights.append(w)
    return values, weights, capacity

def turbo_boosted_knapsack_solver(values, weights, capacity):
    """
    Turbo-boosted Branch and Bound solver for the 0-1 Knapsack Problem.

    Enhancements:
      - Items are sorted by value-to-weight ratio (descending).
      - Precomputed cumulative sums allow a fast (binary search based)
        calculation of a fractional (greedy) upper bound.
      - A best-first search using a priority queue (max-heap) is employed.

    Parameters:
      values: list of item values.
      weights: list of item weights.
      capacity: maximum weight allowed.

    Returns:
      best_value: optimal total value.
      best_solution: binary list (in original order) indicating selected items.
    """
    n = len(values)

    # Sort items by descending ratio (value/weight)
    sorted_indices = sorted(range(n), key=lambda i: values[i] / weights[i], reverse=True)
    sorted_values = [values[i] for i in sorted_indices]
    sorted_weights = [weights[i] for i in sorted_indices]

    # Precompute cumulative sums for the sorted order.
    cumulative_weight = np.zeros(n + 1)
    cumulative_value = np.zeros(n + 1)
    for j in range(n):
        cumulative_weight[j + 1] = cumulative_weight[j] + sorted_weights[j]
        cumulative_value[j + 1] = cumulative_value[j] + sorted_values[j]

    def compute_bound(index, curr_weight, curr_value):
        """
        Compute an upper bound on the total value that can be achieved,
        starting from the current branch (with items [index, n-1] in the sorted order).

        Uses binary search on the precomputed cumulative arrays.
        """
        remaining = capacity - curr_weight
        if remaining <= 0:
            return curr_value
        low = index
        high = n
        while low < high:
            mid = (low + high + 1) // 2
            if cumulative_weight[mid] - cumulative_weight[index] <= remaining:
                low = mid
            else:
                high = mid - 1
        bound_value = curr_value + (cumulative_value[low] - cumulative_value[index])
        if low < n:
            remaining_fraction = remaining - (cumulative_weight[low] - cumulative_weight[index])
            if sorted_weights[low] > 0:
                fraction = remaining_fraction / sorted_weights[low]
                bound_value += fraction * sorted_values[low]
        return bound_value

    # Priority queue node: (negative_bound, current_value, current_weight, index, taken_items)
    initial_bound = compute_bound(0, 0, 0)
    pq = [(-initial_bound, 0, 0, 0, [])]

    best_value = 0
    best_taken_sorted = []  # will store indices in original order

    while pq:
        neg_bound, curr_value, curr_weight, index, taken_sorted = heapq.heappop(pq)
        bound = -neg_bound  # convert back to positive

        if bound <= best_value:
            continue

        if index >= n:
            if curr_value > best_value:
                best_value = curr_value
                best_taken_sorted = taken_sorted
            continue

        # Consider the next item in sorted order.
        item_original = sorted_indices[index]
        item_weight = sorted_weights[index]
        item_value = sorted_values[index]

        # Branch 1: Include the item (if it fits).
        if curr_weight + item_weight <= capacity:
            new_value = curr_value + item_value
            new_weight = curr_weight + item_weight
            new_taken_sorted = taken_sorted + [item_original]
            new_bound = compute_bound(index + 1, new_weight, new_value)
            if new_value > best_value:
                best_value = new_value
                best_taken_sorted = new_taken_sorted
            if new_bound > best_value:
                heapq.heappush(pq, (-new_bound, new_value, new_weight, index + 1, new_taken_sorted))

        # Branch 2: Exclude the item.
        exclude_bound = compute_bound(index + 1, curr_weight, curr_value)
        if exclude_bound > best_value:
            heapq.heappush(pq, (-exclude_bound, curr_value, curr_weight, index + 1, taken_sorted))

    # Reconstruct the solution in the original order.
    best_solution = [0] * n
    for i in best_taken_sorted:
        best_solution[i] = 1

    return best_value, best_solution

def print_table(header, rows, max_lines=20):
    """
    Utility function to print a table in a summarized form if there are many rows.
    If total rows > max_lines, prints first half and last half with an ellipsis in between.
    """
    print(header)
    n = len(rows)
    if n <= max_lines:
        for row in rows:
            print("\t".join(map(str, row)))
    else:
        half = max_lines // 2
        for row in rows[:half]:
            print("\t".join(map(str, row)))
        print("...")
        for row in rows[-half:]:
            print("\t".join(map(str, row)))

def main():
    # File path to the knapsack instance.
    file_path = "/content/knapsack-01-instances-main/pisinger_instances_01_KP/large_scale/knapPI_2_1000_1000_1"

    try:
        values, weights, capacity = load_knapsack_instance(file_path)
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found. Please check the path.")
        return
    except Exception as e:
        print(f"Error loading instance: {e}")
        return

    n = len(values)
    print("Loaded knapsack instance:")
    print(f"  Number of items: {n}")
    print(f"  Capacity: {capacity}\n")

    # Prepare and print the Items table.
    items_rows = [["Index", "Value", "Weight"]]
    for i in range(n):
        items_rows.append([i, values[i], weights[i]])

    print("Items:")
    MAX_LINES_TO_PRINT = 20  # Adjust as needed.
    print_table("Index\tValue\tWeight", items_rows[1:], max_lines=MAX_LINES_TO_PRINT)
    print("-" * 30)

    # Solve using the turbo-boosted solver.
    optimal_value, selected_items = turbo_boosted_knapsack_solver(values, weights, capacity)

    # Double-check: recompute total weight and total value from selected items.
    selected_indices = [i for i, bit in enumerate(selected_items) if bit == 1]
    total_weight = sum(weights[i] for i in selected_indices)
    total_value = sum(values[i] for i in selected_indices)

    print("\nOptimal Solution:")
    print(f"  Optimal Value (from solver): {optimal_value}")
    print(f"  Selected Items (binary vector): {selected_items}")
    print(f"  Selected Item Indices: {selected_indices}")
    print(f"  Total Weight of Selected Items: {total_weight}")
    print(f"  Total Value of Selected Items: {total_value}")

    if total_value != optimal_value:
        print("WARNING: The computed total value does not match the solver's optimal value!")

    # Prepare and print the Detailed Items Selection table.
    detailed_rows = [["Index", "Value", "Weight", "Selected"]]
    for i in range(n):
        selected_flag = "Yes" if selected_items[i] == 1 else "No"
        detailed_rows.append([i, values[i], weights[i], selected_flag])

    print("\nDetailed Items Selection:")
    print_table("Index\tValue\tWeight\tSelected", detailed_rows[1:], max_lines=MAX_LINES_TO_PRINT)

    # Print the final optimal solution summary at the end.
    print("\n" + "=" * 40)
    print("FINAL OPTIMAL SOLUTION SUMMARY")
    print("=" * 40)
    print(f"Optimal Value: {optimal_value}")
    print(f"Selected Items (binary vector): {selected_items}")
    print(f"Selected Item Indices: {selected_indices}")
    print(f"Total Weight: {total_weight}")
    print(f"Total Value: {total_value}")
    print("=" * 40)

if __name__ == "__main__":
    # Open output file and redirect sys.stdout using a try/finally block to ensure restoration.
    out_file = open("knapsack_output.txt", "w")
    original_stdout = sys.stdout
    sys.stdout = Tee(original_stdout, out_file)
    try:
        main()
    finally:
        # Flush and restore the original stdout, then close the file.
        sys.stdout.flush()
        sys.stdout = original_stdout
        out_file.close()
```

```python
import heapq
import numpy as np
import sys

class Tee:
    """
    A helper class that duplicates output to multiple streams.
    """
    def __init__(self, *files):
        self.files = files

    def write(self, data):
        for f in self.files:
            f.write(data)

    def flush(self):
        for f in self.files:
            f.flush()

def load_knapsack_instance(file_path):
    """
    Loads a knapsack instance in the Pisinger format.

    Expected file format:
      - The first line contains two integers: n (number of items) and capacity.
      - The next n lines each contain two integers: value and weight for each item.

    Returns:
      values (list of int): The list of item values.
      weights (list of int): The list of item weights.
      capacity (int): The knapsack capacity.
    """
    with open(file_path, "r") as f:
        lines = f.readlines()

    if not lines:
        raise ValueError("The file is empty.")

    try:
        n, capacity = map(int, lines[0].split())
    except Exception as e:
        raise ValueError("The first line must contain two integers: number of items and capacity.") from e

    if len(lines) < n + 1:
        raise ValueError(f"Expected {n} item lines, but only found {len(lines)-1}.")

    values, weights = [], []
    for i in range(1, n + 1):
        try:
            v, w = map(int, lines[i].split())
        except Exception as e:
            raise ValueError(f"Line {i+1} is not in the expected 'value weight' format.") from e
        values.append(v)
        weights.append(w)
    return values, weights, capacity

def turbo_boosted_knapsack_solver(values, weights, capacity):
    """
    Turbo-boosted Branch and Bound solver for the 0-1 Knapsack Problem.

    Enhancements:
      - Items are sorted by value-to-weight ratio (descending).
      - Precomputed cumulative sums allow a fast (binary search based)
        calculation of a fractional (greedy) upper bound.
      - A best-first search using a priority queue (max-heap) is employed.

    Parameters:
      values: list of item values.
      weights: list of item weights.
      capacity: maximum weight allowed.

    Returns:
      best_value: optimal total value.
      best_solution: binary list (in original order) indicating selected items.
    """
    n = len(values)

    # Sort items by descending ratio (value/weight)
    sorted_indices = sorted(range(n), key=lambda i: values[i] / weights[i], reverse=True)
    sorted_values = [values[i] for i in sorted_indices]
    sorted_weights = [weights[i] for i in sorted_indices]

    # Precompute cumulative sums for the sorted order.
    cumulative_weight = np.zeros(n + 1)
    cumulative_value = np.zeros(n + 1)
    for j in range(n):
        cumulative_weight[j + 1] = cumulative_weight[j] + sorted_weights[j]
        cumulative_value[j + 1] = cumulative_value[j] + sorted_values[j]

    def compute_bound(index, curr_weight, curr_value):
        """
        Compute an upper bound on the total value that can be achieved,
        starting from the current branch (with items [index, n-1] in the sorted order).

        Uses binary search on the precomputed cumulative arrays.
        """
        remaining = capacity - curr_weight
        if remaining <= 0:
            return curr_value
        low = index
        high = n
        while low < high:
            mid = (low + high + 1) // 2
            if cumulative_weight[mid] - cumulative_weight[index] <= remaining:
                low = mid
            else:
                high = mid - 1
        bound_value = curr_value + (cumulative_value[low] - cumulative_value[index])
        if low < n:
            remaining_fraction = remaining - (cumulative_weight[low] - cumulative_weight[index])
            if sorted_weights[low] > 0:
                fraction = remaining_fraction / sorted_weights[low]
                bound_value += fraction * sorted_values[low]
        return bound_value

    # Priority queue node: (negative_bound, current_value, current_weight, index, taken_items)
    initial_bound = compute_bound(0, 0, 0)
    pq = [(-initial_bound, 0, 0, 0, [])]

    best_value = 0
    best_taken_sorted = []  # will store indices in original order

    while pq:
        neg_bound, curr_value, curr_weight, index, taken_sorted = heapq.heappop(pq)
        bound = -neg_bound  # convert back to positive

        if bound <= best_value:
            continue

        if index >= n:
            if curr_value > best_value:
                best_value = curr_value
                best_taken_sorted = taken_sorted
            continue

        # Consider the next item in sorted order.
        item_original = sorted_indices[index]
        item_weight = sorted_weights[index]
        item_value = sorted_values[index]

        # Branch 1: Include the item (if it fits).
        if curr_weight + item_weight <= capacity:
            new_value = curr_value + item_value
            new_weight = curr_weight + item_weight
            new_taken_sorted = taken_sorted + [item_original]
            new_bound = compute_bound(index + 1, new_weight, new_value)
            if new_value > best_value:
                best_value = new_value
                best_taken_sorted = new_taken_sorted
            if new_bound > best_value:
                heapq.heappush(pq, (-new_bound, new_value, new_weight, index + 1, new_taken_sorted))

        # Branch 2: Exclude the item.
        exclude_bound = compute_bound(index + 1, curr_weight, curr_value)
        if exclude_bound > best_value:
            heapq.heappush(pq, (-exclude_bound, curr_value, curr_weight, index + 1, taken_sorted))

    # Reconstruct the solution in the original order.
    best_solution = [0] * n
    for i in best_taken_sorted:
        best_solution[i] = 1

    return best_value, best_solution

def print_table(header, rows, max_lines=20):
    """
    Utility function to print a table in a summarized form if there are many rows.
    If total rows > max_lines, prints first half and last half with an ellipsis in between.
    """
    print(header)
    n = len(rows)
    if n <= max_lines:
        for row in rows:
            print("\t".join(map(str, row)))
    else:
        half = max_lines // 2
        for row in rows[:half]:
            print("\t".join(map(str, row)))
        print("...")
        for row in rows[-half:]:
            print("\t".join(map(str, row)))

def main():
    # File path to the knapsack instance.
    file_path = "/content/knapsack-01-instances-main/pisinger_instances_01_KP/large_scale/knapPI_2_500_1000_1"

    try:
        values, weights, capacity = load_knapsack_instance(file_path)
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found. Please check the path.")
        return
    except Exception as e:
        print(f"Error loading instance: {e}")
        return

    n = len(values)
    print("Loaded knapsack instance:")
    print(f"  Number of items: {n}")
    print(f"  Capacity: {capacity}\n")

    # Prepare and print the Items table.
    items_rows = [["Index", "Value", "Weight"]]
    for i in range(n):
        items_rows.append([i, values[i], weights[i]])

    print("Items:")
    MAX_LINES_TO_PRINT = 20  # Adjust as needed.
    print_table("Index\tValue\tWeight", items_rows[1:], max_lines=MAX_LINES_TO_PRINT)
    print("-" * 30)

    # Solve using the turbo-boosted solver.
    optimal_value, selected_items = turbo_boosted_knapsack_solver(values, weights, capacity)

    # Double-check: recompute total weight and total value from selected items.
    selected_indices = [i for i, bit in enumerate(selected_items) if bit == 1]
    total_weight = sum(weights[i] for i in selected_indices)
    total_value = sum(values[i] for i in selected_indices)

    print("\nOptimal Solution:")
    print(f"  Optimal Value (from solver): {optimal_value}")
    print(f"  Selected Items (binary vector): {selected_items}")
    print(f"  Selected Item Indices: {selected_indices}")
    print(f"  Total Weight of Selected Items: {total_weight}")
    print(f"  Total Value of Selected Items: {total_value}")

    if total_value != optimal_value:
        print("WARNING: The computed total value does not match the solver's optimal value!")

    # Prepare and print the Detailed Items Selection table.
    detailed_rows = [["Index", "Value", "Weight", "Selected"]]
    for i in range(n):
        selected_flag = "Yes" if selected_items[i] == 1 else "No"
        detailed_rows.append([i, values[i], weights[i], selected_flag])

    print("\nDetailed Items Selection:")
    print_table("Index\tValue\tWeight\tSelected", detailed_rows[1:], max_lines=MAX_LINES_TO_PRINT)

    # Print the final optimal solution summary at the end.
    print("\n" + "=" * 40)
    print("FINAL OPTIMAL SOLUTION SUMMARY")
    print("=" * 40)
    print(f"Optimal Value: {optimal_value}")
    print(f"Selected Items (binary vector): {selected_items}")
    print(f"Selected Item Indices: {selected_indices}")
    print(f"Total Weight: {total_weight}")
    print(f"Total Value: {total_value}")
    print("=" * 40)

if __name__ == "__main__":
    # Open output file and redirect sys.stdout using a try/finally block to ensure restoration.
    out_file = open("knapsack_output.txt", "w")
    original_stdout = sys.stdout
    sys.stdout = Tee(original_stdout, out_file)
    try:
        main()
    finally:
        # Flush and restore the original stdout, then close the file.
        sys.stdout.flush()
        sys.stdout = original_stdout
        out_file.close()
```

```python
import heapq
import numpy as np
import sys

class Tee:
    """
    A helper class that duplicates output to multiple streams.
    """
    def __init__(self, *files):
        self.files = files

    def write(self, data):
        for f in self.files:
            f.write(data)

    def flush(self):
        for f in self.files:
            f.flush()

def load_knapsack_instance(file_path):
    """
    Loads a knapsack instance in the Pisinger format.

    Expected file format:
      - The first line contains two integers: n (number of items) and capacity.
      - The next n lines each contain two integers: value and weight for each item.

    Returns:
      values (list of int): The list of item values.
      weights (list of int): The list of item weights.
      capacity (int): The knapsack capacity.
    """
    with open(file_path, "r") as f:
        lines = f.readlines()

    if not lines:
        raise ValueError("The file is empty.")

    try:
        n, capacity = map(int, lines[0].split())
    except Exception as e:
        raise ValueError("The first line must contain two integers: number of items and capacity.") from e

    if len(lines) < n + 1:
        raise ValueError(f"Expected {n} item lines, but only found {len(lines)-1}.")

    values, weights = [], []
    for i in range(1, n + 1):
        try:
            v, w = map(int, lines[i].split())
        except Exception as e:
            raise ValueError(f"Line {i+1} is not in the expected 'value weight' format.") from e
        values.append(v)
        weights.append(w)
    return values, weights, capacity

def turbo_boosted_knapsack_solver(values, weights, capacity):
    """
    Turbo-boosted Branch and Bound solver for the 0-1 Knapsack Problem.

    Enhancements:
      - Items are sorted by value-to-weight ratio (descending).
      - Precomputed cumulative sums allow a fast (binary search based)
        calculation of a fractional (greedy) upper bound.
      - A best-first search using a priority queue (max-heap) is employed.

    Parameters:
      values: list of item values.
      weights: list of item weights.
      capacity: maximum weight allowed.

    Returns:
      best_value: optimal total value.
      best_solution: binary list (in original order) indicating selected items.
    """
    n = len(values)

    # Sort items by descending ratio (value/weight)
    sorted_indices = sorted(range(n), key=lambda i: values[i] / weights[i], reverse=True)
    sorted_values = [values[i] for i in sorted_indices]
    sorted_weights = [weights[i] for i in sorted_indices]

    # Precompute cumulative sums for the sorted order.
    cumulative_weight = np.zeros(n + 1)
    cumulative_value = np.zeros(n + 1)
    for j in range(n):
        cumulative_weight[j + 1] = cumulative_weight[j] + sorted_weights[j]
        cumulative_value[j + 1] = cumulative_value[j] + sorted_values[j]

    def compute_bound(index, curr_weight, curr_value):
        """
        Compute an upper bound on the total value that can be achieved,
        starting from the current branch (with items [index, n-1] in the sorted order).

        Uses binary search on the precomputed cumulative arrays.
        """
        remaining = capacity - curr_weight
        if remaining <= 0:
            return curr_value
        low = index
        high = n
        while low < high:
            mid = (low + high + 1) // 2
            if cumulative_weight[mid] - cumulative_weight[index] <= remaining:
                low = mid
            else:
                high = mid - 1
        bound_value = curr_value + (cumulative_value[low] - cumulative_value[index])
        if low < n:
            remaining_fraction = remaining - (cumulative_weight[low] - cumulative_weight[index])
            if sorted_weights[low] > 0:
                fraction = remaining_fraction / sorted_weights[low]
                bound_value += fraction * sorted_values[low]
        return bound_value

    # Priority queue node: (negative_bound, current_value, current_weight, index, taken_items)
    initial_bound = compute_bound(0, 0, 0)
    pq = [(-initial_bound, 0, 0, 0, [])]

    best_value = 0
    best_taken_sorted = []  # will store indices in original order

    while pq:
        neg_bound, curr_value, curr_weight, index, taken_sorted = heapq.heappop(pq)
        bound = -neg_bound  # convert back to positive

        if bound <= best_value:
            continue

        if index >= n:
            if curr_value > best_value:
                best_value = curr_value
                best_taken_sorted = taken_sorted
            continue

        # Consider the next item in sorted order.
        item_original = sorted_indices[index]
        item_weight = sorted_weights[index]
        item_value = sorted_values[index]

        # Branch 1: Include the item (if it fits).
        if curr_weight + item_weight <= capacity:
            new_value = curr_value + item_value
            new_weight = curr_weight + item_weight
            new_taken_sorted = taken_sorted + [item_original]
            new_bound = compute_bound(index + 1, new_weight, new_value)
            if new_value > best_value:
                best_value = new_value
                best_taken_sorted = new_taken_sorted
            if new_bound > best_value:
                heapq.heappush(pq, (-new_bound, new_value, new_weight, index + 1, new_taken_sorted))

        # Branch 2: Exclude the item.
        exclude_bound = compute_bound(index + 1, curr_weight, curr_value)
        if exclude_bound > best_value:
            heapq.heappush(pq, (-exclude_bound, curr_value, curr_weight, index + 1, taken_sorted))

    # Reconstruct the solution in the original order.
    best_solution = [0] * n
    for i in best_taken_sorted:
        best_solution[i] = 1

    return best_value, best_solution

def print_table(header, rows, max_lines=20):
    """
    Utility function to print a table in a summarized form if there are many rows.
    If total rows > max_lines, prints first half and last half with an ellipsis in between.
    """
    print(header)
    n = len(rows)
    if n <= max_lines:
        for row in rows:
            print("\t".join(map(str, row)))
    else:
        half = max_lines // 2
        for row in rows[:half]:
            print("\t".join(map(str, row)))
        print("...")
        for row in rows[-half:]:
            print("\t".join(map(str, row)))

def main():
    # File path to the knapsack instance.
    file_path = "/content/knapsack-01-instances-main/pisinger_instances_01_KP/large_scale/knapPI_3_200_1000_1"

    try:
        values, weights, capacity = load_knapsack_instance(file_path)
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found. Please check the path.")
        return
    except Exception as e:
        print(f"Error loading instance: {e}")
        return

    n = len(values)
    print("Loaded knapsack instance:")
    print(f"  Number of items: {n}")
    print(f"  Capacity: {capacity}\n")

    # Prepare and print the Items table.
    items_rows = [["Index", "Value", "Weight"]]
    for i in range(n):
        items_rows.append([i, values[i], weights[i]])

    print("Items:")
    MAX_LINES_TO_PRINT = 20  # Adjust as needed.
    print_table("Index\tValue\tWeight", items_rows[1:], max_lines=MAX_LINES_TO_PRINT)
    print("-" * 30)

    # Solve using the turbo-boosted solver.
    optimal_value, selected_items = turbo_boosted_knapsack_solver(values, weights, capacity)

    # Double-check: recompute total weight and total value from selected items.
    selected_indices = [i for i, bit in enumerate(selected_items) if bit == 1]
    total_weight = sum(weights[i] for i in selected_indices)
    total_value = sum(values[i] for i in selected_indices)

    print("\nOptimal Solution:")
    print(f"  Optimal Value (from solver): {optimal_value}")
    print(f"  Selected Items (binary vector): {selected_items}")
    print(f"  Selected Item Indices: {selected_indices}")
    print(f"  Total Weight of Selected Items: {total_weight}")
    print(f"  Total Value of Selected Items: {total_value}")

    if total_value != optimal_value:
        print("WARNING: The computed total value does not match the solver's optimal value!")

    # Prepare and print the Detailed Items Selection table.
    detailed_rows = [["Index", "Value", "Weight", "Selected"]]
    for i in range(n):
        selected_flag = "Yes" if selected_items[i] == 1 else "No"
        detailed_rows.append([i, values[i], weights[i], selected_flag])

    print("\nDetailed Items Selection:")
    print_table("Index\tValue\tWeight\tSelected", detailed_rows[1:], max_lines=MAX_LINES_TO_PRINT)

    # Print the final optimal solution summary at the end.
    print("\n" + "=" * 40)
    print("FINAL OPTIMAL SOLUTION SUMMARY")
    print("=" * 40)
    print(f"Optimal Value: {optimal_value}")
    print(f"Selected Items (binary vector): {selected_items}")
    print(f"Selected Item Indices: {selected_indices}")
    print(f"Total Weight: {total_weight}")
    print(f"Total Value: {total_value}")
    print("=" * 40)

if __name__ == "__main__":
    # Open output file and redirect sys.stdout using a try/finally block to ensure restoration.
    out_file = open("knapsack_output.txt", "w")
    original_stdout = sys.stdout
    sys.stdout = Tee(original_stdout, out_file)
    try:
        main()
    finally:
        # Flush and restore the original stdout, then close the file.
        sys.stdout.flush()
        sys.stdout = original_stdout
        out_file.close()
```

```python
import heapq
import numpy as np
import sys

class Tee:
    """
    A helper class that duplicates output to multiple streams.
    """
    def __init__(self, *files):
        self.files = files

    def write(self, data):
        for f in self.files:
            f.write(data)

    def flush(self):
        for f in self.files:
            f.flush()

def load_knapsack_instance(file_path):
    """
    Loads a knapsack instance in the Pisinger format.

    Expected file format:
      - The first line contains two integers: n (number of items) and capacity.
      - The next n lines each contain two integers: value and weight for each item.

    Returns:
      values (list of int): The list of item values.
      weights (list of int): The list of item weights.
      capacity (int): The knapsack capacity.
    """
    with open(file_path, "r") as f:
        lines = f.readlines()

    if not lines:
        raise ValueError("The file is empty.")

    try:
        n, capacity = map(int, lines[0].split())
    except Exception as e:
        raise ValueError("The first line must contain two integers: number of items and capacity.") from e

    if len(lines) < n + 1:
        raise ValueError(f"Expected {n} item lines, but only found {len(lines)-1}.")

    values, weights = [], []
    for i in range(1, n + 1):
        try:
            v, w = map(int, lines[i].split())
        except Exception as e:
            raise ValueError(f"Line {i+1} is not in the expected 'value weight' format.") from e
        values.append(v)
        weights.append(w)
    return values, weights, capacity

def turbo_boosted_knapsack_solver(values, weights, capacity):
    """
    Turbo-boosted Branch and Bound solver for the 0-1 Knapsack Problem.

    Enhancements:
      - Items are sorted by value-to-weight ratio (descending).
      - Precomputed cumulative sums allow a fast (binary search based)
        calculation of a fractional (greedy) upper bound.
      - A best-first search using a priority queue (max-heap) is employed.

    Parameters:
      values: list of item values.
      weights: list of item weights.
      capacity: maximum weight allowed.

    Returns:
      best_value: optimal total value.
      best_solution: binary list (in original order) indicating selected items.
    """
    n = len(values)

    # Sort items by descending ratio (value/weight)
    sorted_indices = sorted(range(n), key=lambda i: values[i] / weights[i], reverse=True)
    sorted_values = [values[i] for i in sorted_indices]
    sorted_weights = [weights[i] for i in sorted_indices]

    # Precompute cumulative sums for the sorted order.
    cumulative_weight = np.zeros(n + 1)
    cumulative_value = np.zeros(n + 1)
    for j in range(n):
        cumulative_weight[j + 1] = cumulative_weight[j] + sorted_weights[j]
        cumulative_value[j + 1] = cumulative_value[j] + sorted_values[j]

    def compute_bound(index, curr_weight, curr_value):
        """
        Compute an upper bound on the total value that can be achieved,
        starting from the current branch (with items [index, n-1] in the sorted order).

        Uses binary search on the precomputed cumulative arrays.
        """
        remaining = capacity - curr_weight
        if remaining <= 0:
            return curr_value
        low = index
        high = n
        while low < high:
            mid = (low + high + 1) // 2
            if cumulative_weight[mid] - cumulative_weight[index] <= remaining:
                low = mid
            else:
                high = mid - 1
        bound_value = curr_value + (cumulative_value[low] - cumulative_value[index])
        if low < n:
            remaining_fraction = remaining - (cumulative_weight[low] - cumulative_weight[index])
            if sorted_weights[low] > 0:
                fraction = remaining_fraction / sorted_weights[low]
                bound_value += fraction * sorted_values[low]
        return bound_value

    # Priority queue node: (negative_bound, current_value, current_weight, index, taken_items)
    initial_bound = compute_bound(0, 0, 0)
    pq = [(-initial_bound, 0, 0, 0, [])]

    best_value = 0
    best_taken_sorted = []  # will store indices in original order

    while pq:
        neg_bound, curr_value, curr_weight, index, taken_sorted = heapq.heappop(pq)
        bound = -neg_bound  # convert back to positive

        if bound <= best_value:
            continue

        if index >= n:
            if curr_value > best_value:
                best_value = curr_value
                best_taken_sorted = taken_sorted
            continue

        # Consider the next item in sorted order.
        item_original = sorted_indices[index]
        item_weight = sorted_weights[index]
        item_value = sorted_values[index]

        # Branch 1: Include the item (if it fits).
        if curr_weight + item_weight <= capacity:
            new_value = curr_value + item_value
            new_weight = curr_weight + item_weight
            new_taken_sorted = taken_sorted + [item_original]
            new_bound = compute_bound(index + 1, new_weight, new_value)
            if new_value > best_value:
                best_value = new_value
                best_taken_sorted = new_taken_sorted
            if new_bound > best_value:
                heapq.heappush(pq, (-new_bound, new_value, new_weight, index + 1, new_taken_sorted))

        # Branch 2: Exclude the item.
        exclude_bound = compute_bound(index + 1, curr_weight, curr_value)
        if exclude_bound > best_value:
            heapq.heappush(pq, (-exclude_bound, curr_value, curr_weight, index + 1, taken_sorted))

    # Reconstruct the solution in the original order.
    best_solution = [0] * n
    for i in best_taken_sorted:
        best_solution[i] = 1

    return best_value, best_solution

def print_table(header, rows, max_lines=20):
    """
    Utility function to print a table in a summarized form if there are many rows.
    If total rows > max_lines, prints first half and last half with an ellipsis in between.
    """
    print(header)
    n = len(rows)
    if n <= max_lines:
        for row in rows:
            print("\t".join(map(str, row)))
    else:
        half = max_lines // 2
        for row in rows[:half]:
            print("\t".join(map(str, row)))
        print("...")
        for row in rows[-half:]:
            print("\t".join(map(str, row)))

def main():
    # File path to the knapsack instance.
    file_path = "/content/knapsack-01-instances-main/pisinger_instances_01_KP/large_scale/knapPI_3_500_1000_1"

    try:
        values, weights, capacity = load_knapsack_instance(file_path)
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found. Please check the path.")
        return
    except Exception as e:
        print(f"Error loading instance: {e}")
        return

    n = len(values)
    print("Loaded knapsack instance:")
    print(f"  Number of items: {n}")
    print(f"  Capacity: {capacity}\n")

    # Prepare and print the Items table.
    items_rows = [["Index", "Value", "Weight"]]
    for i in range(n):
        items_rows.append([i, values[i], weights[i]])

    print("Items:")
    MAX_LINES_TO_PRINT = 20  # Adjust as needed.
    print_table("Index\tValue\tWeight", items_rows[1:], max_lines=MAX_LINES_TO_PRINT)
    print("-" * 30)

    # Solve using the turbo-boosted solver.
    optimal_value, selected_items = turbo_boosted_knapsack_solver(values, weights, capacity)

    # Double-check: recompute total weight and total value from selected items.
    selected_indices = [i for i, bit in enumerate(selected_items) if bit == 1]
    total_weight = sum(weights[i] for i in selected_indices)
    total_value = sum(values[i] for i in selected_indices)

    print("\nOptimal Solution:")
    print(f"  Optimal Value (from solver): {optimal_value}")
    print(f"  Selected Items (binary vector): {selected_items}")
    print(f"  Selected Item Indices: {selected_indices}")
    print(f"  Total Weight of Selected Items: {total_weight}")
    print(f"  Total Value of Selected Items: {total_value}")

    if total_value != optimal_value:
        print("WARNING: The computed total value does not match the solver's optimal value!")

    # Prepare and print the Detailed Items Selection table.
    detailed_rows = [["Index", "Value", "Weight", "Selected"]]
    for i in range(n):
        selected_flag = "Yes" if selected_items[i] == 1 else "No"
        detailed_rows.append([i, values[i], weights[i], selected_flag])

    print("\nDetailed Items Selection:")
    print_table("Index\tValue\tWeight\tSelected", detailed_rows[1:], max_lines=MAX_LINES_TO_PRINT)

    # Print the final optimal solution summary at the end.
    print("\n" + "=" * 40)
    print("FINAL OPTIMAL SOLUTION SUMMARY")
    print("=" * 40)
    print(f"Optimal Value: {optimal_value}")
    print(f"Selected Items (binary vector): {selected_items}")
    print(f"Selected Item Indices: {selected_indices}")
    print(f"Total Weight: {total_weight}")
    print(f"Total Value: {total_value}")
    print("=" * 40)

if __name__ == "__main__":
    # Open output file and redirect sys.stdout using a try/finally block to ensure restoration.
    out_file = open("knapsack_output.txt", "w")
    original_stdout = sys.stdout
    sys.stdout = Tee(original_stdout, out_file)
    try:
        main()
    finally:
        # Flush and restore the original stdout, then close the file.
        sys.stdout.flush()
        sys.stdout = original_stdout
        out_file.close()
```
